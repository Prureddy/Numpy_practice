```python
#random
from numpy import random
x = random.randint(50,100,size=10)
print(x)

#randint()

x = random.randint(100,size=(3,3,3))
print(x)
```

```python
#rand()

x = random.rand(3,5)
print(x)
```

```python
#choice

x = random.choice([1,2,3,4,5,6,7,8,6])
print(x)
```

```python
#Random Data Distribution
from numpy import random

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.5, 0.1], size=(3, 5))

print(x)
```

```python
#shuffle()
import numpy as np
from numpy import random
arr = np.array([1,2,3,4,5,6])
print(arr)
random.shuffle(arr)
print(arr)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.normal(loc=1, scale=2, size=(4,3))

sns.kdeplot(x)
sns.displot(x)
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.binomial(n=10, p=0.5, size=1000),kde=False)

plt.show()
```

```python
#Difference Between Normal and Binomial Distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


random_data = np.random.normal(loc=50, scale=5, size=1000)
binomial_data = np.random.binomial(n=100, p=0.5, size=1000)

sns.kdeplot(random_data,label='normal')
sns.kdeplot(binomial_data,label='binomial')

plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
normal_data = np.random.normal(loc=50, scale=5, size=1000)
poisson_data = np.random.poisson(lam=50, size=1000)

# Create kernel density estimate plot using kdeplot
sns.kdeplot(normal_data, label='normal')
sns.kdeplot(poisson_data, label='poisson')

# Display the plot
plt.legend()
plt.show()
```

```python
#exponential distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

expo=np.random.exponential(size=1000)
normal_data = np.random.normal(loc=5, scale=0.5, size=1000)
sns.kdeplot(expo,label='exponential')
sns.kdeplot(normal_data, label='normal')

plt.legend()
plt.show()
```
